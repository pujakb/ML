```python
import pandas as pd
import numpy as np
import collections
import matplotlib.pyplot as plt
from matplotlib import colors
```

```python
df = pd.read_csv(r'C:\Users\WineQA.csv', header=None)
X = df.iloc[1:500, [0,10]].values
```

```python
df.head(5)

```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3,init='random',n_init=10,max_iter=300,tol=1e-04,random_state=0)
y_km = km.fit_predict(X)
```

```python
# plt.scatter(X[y_km==0,0],X[y_km==0,1],s=50, c='blue',marker='s',label='cluster 1')
# plt.scatter(X[y_km==1,0],X[y_km==1,1],s=50,c='orange',marker='o',label='cluster 2')
# plt.scatter(X[y_km==2,0],X[y_km==2,1],s=50,c='black',marker='x',label='cluster 3')
color_map=['black', 'lime', 'red', 'yellow', 'blue', 'green']
plt.scatter(X[:, 0], X[:, 1], c=data_labels, s=20, cmap=colors.ListedColormap(color_map))
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=250,marker='*',c='red',label='centroids')
# plt.ylim(5,50)
# plt.xlim(5,50)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c='red',marker='x')
plt.show()
```

```python
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i,
        init='k-means++',
        n_init=10,
        max_iter=300,
        random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)
plt.plot(range(1,11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```
